***Simple programm in Qiskit***

```python
!pip install qiskit
import qiskit
```

creat a quantum rgister with one qubit


```python
qreg=qiskit.QuantumRegister(1,name='qreg')
```

creat a classical register with one qubit

```python
creg=qiskit.ClassicalRegister(1,name='creg')
```

Q Circuit

```python
circ=qiskit.QuantumCircuit(qreg,creg)
```

Add Not operation on the qubit

```python
circ.x(qreg[0])
```

Add a measurement on the qubit

```python
circ.measure(qreg,creg)
```

print the circuit

```python
print(circ.draw())
```

get a backend to run on

```python
!pip install qiskit-aer
from qiskit_aer import AerSimulator
backend=AerSimulator()
```

```python
job=backend.run(circ,shots=10)
result=job.result()
```

```python
print(result.get_counts())
```
